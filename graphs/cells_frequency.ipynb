```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load JSON data
with open('../data/count.json') as f:
    data = json.load(f)

# Create an empty 8x8 array
cells = np.zeros((8, 8))

# Fill the array with the values from the JSON data
for key, value in data.items():
    row = ord(key[0]) - ord('a')
    col = int(key[1]) - 1
    cells[row, col] = value
```

```python
# Plotting
plt.imshow(cells, cmap='plasma', origin='lower', interpolation='nearest', extent=[0, 8, 0, 8])
plt.colorbar(label='Occurrences')

# Manually setting x-axis ticks and labels
plt.xticks(np.arange(0.5, 8.5, 1), [chr(ord('a') + i) for i in range(8)])
plt.xlabel('Column')

# Manually setting y-axis ticks and labels
plt.yticks(np.arange(0.5, 8.5, 1), [str(i+1) for i in range(8)])
plt.ylabel('Row')

plt.title('Heat map of moves played')
plt.show()
```
