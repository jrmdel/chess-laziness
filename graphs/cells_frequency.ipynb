```python
# Import libraries

import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define functions

def load_json_file(path):
  with open(path) as f:
    data = json.load(f)
    return data

def convert_json_to_cells(input):
  cells = np.zeros((8, 8))

  for key, value in input.items():
    col = ord(key[0]) - ord('a')
    row = int(key[1]) - 1
    cells[row, col] = value
  
  return cells

def plot_heatmap(cells, title):
  plt.imshow(cells, cmap='magma', origin='lower', interpolation='nearest', extent=[0, 8, 0, 8])
  plt.colorbar(label='Occurrences')

  plt.xticks(np.arange(0.5, 8.5, 1), [chr(ord('a') + i) for i in range(8)])
  plt.xlabel('File')

  plt.yticks(np.arange(0.5, 8.5, 1), [str(i+1) for i in range(8)])
  plt.ylabel('Rank')

  plt.title(title)
  plt.show()
```

```python
# All the moves

input = load_json_file(path='../data/cell-count/all.json')
cells = convert_json_to_cells(input)
plot_heatmap(cells, 'Heat map of moves played')
```

```python
# Knight moves

input = load_json_file(path='../data/cell-count/knight.json')
cells = convert_json_to_cells(input)
plot_heatmap(cells, 'Moves made by knights')
```
