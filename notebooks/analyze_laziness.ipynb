```python
# Load libraries

import pandas as pd
import prince
```

```python
df = pd.read_csv('output.csv')
```

```python
indexed_df = df.set_index(['index', 'opening', 'date', 'player_name', 'opponent_name']).drop(['eco', 'event'], axis=1)
```

```python
# narrowed_df = df.loc[df['color'] == 1, ['index', 'color', 'result', 'moves', 'rating_difference', 'distance', 'distance_difference', 'effort', 'effort_difference']]

pca = prince.PCA(
    n_components=3,
    n_iter=3,
    copy=True,
    check_input=True,
    engine='sklearn',
    random_state=42
)
pca = pca.fit(indexed_df)
```

```python
pca.plot(
    indexed_df,
    x_component=0,
    y_component=1,
    color_rows_by='player_name',
    show_row_markers=True,
    show_column_markers=True,
    show_row_labels=False,
    show_column_labels=False
)
```
